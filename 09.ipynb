```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
dir = 'C:/Users/ADMIN/Pictures/Saved Pictures/trump.png'
```

```python
im = Image.open(dir)
plt.imshow(im)
```

```python
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=100)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
