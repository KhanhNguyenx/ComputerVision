```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
image_path = 'C:/Users/ADMIN/Pictures/Saved Pictures/trump.png'
image = Image.open(image_path)
gray_image = image.convert('L')
image_a = np.array(gray_image)
gradient_x = ndimage.convolve(image_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
gradient_y = ndimage.convolve(image_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
fig, axs = plt.subplots(2,1, figsize=(10, 8))
plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('nh xm gc')
plt.subplot(2,1,2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('nh xm gc')
plt.show()
```
